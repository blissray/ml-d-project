```python
with open("./data/farm/farm-ads", "r") as file:
    data = file.readlines()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
cv.fit_transform(data).toarray()
```

```python
cv.vocabulary_
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(max_df=0.3, min_df=0.2)
tf_result = tfv.fit_transform(data).toarray()
tf_result
```
