```
from matplotlib.pyplot import imread
import imgaug as ia
import imgaug.augmenters as iaa
from PIL import Image

target_dir = os.path.join(DATA_DIR, TRAIN_DIR, "dog")

for filename in os.listdir(target_dir):
  target_file_path = os.path.join(target_dir, filename)
  image = imread(target_file_path)
  seq = iaa.Sequential([
      iaa.Crop(px=(0, 16)), # crop images from each side by 0 to 16px (randomly chosen)
      iaa.Fliplr(0.5), # horizontally flip 50% of the images
      iaa.GaussianBlur(sigma=(0, 3.0)) # blur images with a sigma of 0 to 3.0
  ])
  for _ in range(50):
      images_aug = seq.augment_images(image)
      im = Image.fromarray(images_aug)
      im.save(os.path.join(target_dir, "img_aug_{}".format(_) + filename))
```

```
def _preprocess_image(image):   
    image = tf.image.decode_jpeg(image)   
    image = tf.image.convert_image_dtype(image, tf.float32)
    image = tf.image.resize(image, [128, 128])
    # image /= 255.0  # normalize to [0,1] range    
    return image

def _load_and_preprocess_image(path):   
    image = tf.io.read_file(path)
    image = _preprocess_image(image)
    return image

def load_and_preprocess_from_path_label(path, label):   
    return _load_and_preprocess_image(path), label  

def get_dataset(ds):
    image_label_ds = ds.map(load_and_preprocess_from_path_label)
    image_label_ds = image_label_ds.shuffle(1024)
    image_label_ds = image_label_ds.batch(1024)
    image_label_ds = image_label_ds.prefetch(buffer_size=1024)
    return image_label_ds
```
